```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
without_delay = pd.read_excel("data/dataset.xlsx", sheet_name=0)
without_delay.head()
```

```python
s = f"Сера рент-флу метод; {without_delay.iloc[0, 2]}; {without_delay.iloc[1, 2].capitalize()}"
s
```

```python
without_delay = without_delay.rename(columns={"сера рент-флу метод": s})
without_delay = without_delay.drop(index=[0, 1, 2])
without_delay.head()
```

```python
without_delay.info()
```

```python
without_delay = without_delay.drop(["Unnamed: 0"], axis = 1)
```

```python
without_delay.isna().sum()
```

```python
numeric_df = without_delay.apply(pd.to_numeric, errors = "coerce")
```

```python
numeric_df.isna().sum()
```

```python
numeric_df = numeric_df.dropna()
numeric_df
```

```python
target = "Сера рент-флу метод; Среднее сера (мг/кг); По норме менее 10"
features = without_delay.columns[3:]
fig, axs = plt.subplots(4, 4, figsize = (20, 20))


for ax, feature in zip(axs.flatten(), features):
    sns.scatterplot(data = numeric_df, x = feature, y = target, ax=ax)
    ax.set_title(feature)
    ax.set_xlabel("")
    ax.set_ylabel("")

plt.tight_layout()
plt.show()
```
